# Data Descriptions 
1. Data Name : Penduduk Miskin dan Indeks Kemiskinan
2. Source : https://data.go.id/dataset/penduduk-miskin-dan-indeks-kemiskinan
3. Level : Kabupaten Kota
4. Year : 2007 - 2010
5. Cakupan : Nasional

#### Penduduk Miskin dan Indeks Kemiskinan

Dataset ini berisi mengenai jumlah dan persentase penduduk miskin per kabupaten/kota beserta indeks kemiskinan dan perhitungan garis kemiskinan

#### Penjelasan mengenai Variabel pada Dataset ini:

- kode_provinsi: Kode Provinsi 
- nama_provinsi: Nama Provinsi
- kode_kabkota: Kode Kabupaten/Kota
- nama_kabkota: Nama Kabupaten Kota
- tahun: Tahun
- jumlah_penduduk_miskin: Jumlah Penduduk Miskin
- persen_penduduk_miskin: Persen Penduduk Miskin
- indeks_p1: Indeks P1 (Kedalaman Kemiskinan)
- indeks_p2: Indeks P2 (Keparahan Kemiskinan)
- garis_kemiskinan: Garis Kemiskinan (Rupiah/Kapita/bulan)

#### Note:
- Namun pada analisis ini, dilakukan drop atribut (kode_provinsi, kode_kabkota, indeks_p1 dan indeks_p2).
- Hanya digunakan atribut (nama_provinsi, nama_kabkota, jumlah_penduduk_miskin, persen_penduduk_miskin dan garis_kemiskinan)

# Visualisations

```python
# Grafik Rata-Rata Persen Penduduk Miskin >30 Tahun 2009

vis = data_penduduk[(data_penduduk['tahun']==2009) & (data_penduduk['persen_penduduk_miskin']>30) ].groupby('nama_provinsi').mean()['persen_penduduk_miskin']
# print (vis)
vis = vis.plot(kind='pie', figsize=(6, 6),
                            autopct='%1.1f%%', # add in percentages
                            startangle=90,     # start angle 90Â°
                            shadow=True, 
#                             pctdistance=1.12,  
                            colors=['lightblue','blue','yellow','pink','lightgreen','red'],
title ="Rata-Rata Persen Penduduk Miskin >30 Tahun 2009", y=1.2)
```

```python
# Grafik Total Garis Kemiskinan Periode 2007-2010 di Jawa Timur, Papua Barat, dan Sumatera Utara

vis = data_penduduk[(data_penduduk['nama_provinsi']=='Prov. Sumatera Utara') | (data_penduduk['nama_provinsi']=='Prov. Papua Barat')| (data_penduduk['nama_provinsi']=='Prov. Jawa Timur')].groupby('nama_provinsi').sum()['garis_kemiskinan']
# print (vis)
vis = vis.plot(kind='bar', 
               title ="Total Garis Kemiskinan Periode 2007-2010 \n (Jawa Timur, Papua Barat, dan Sumatera Utara)", 
               figsize=(10,5),
               align='center', alpha=0.7
              )
vis.set_xlabel("Provinsi", fontsize=12)
vis.set_ylabel("Total Garis Kemiskinan", fontsize=12)
```

# Analysis 

# 1. Analisis Visualisasi Gambar 1
Hasil visualisasi pada gambar 1 menunjukkan bahwa terdapat 6 Provinsi yang nilai rata-rata persen penduduk miskin besar dari 30% di tahun 2009. Dari 6 Provinsi tersebut nilai terbesar dimiliki oleh Provinsi Bengkulu sebesar 20%, yang kemudian diikuti Provinsi Papua (18.9%), Provinsi Papua Barat(18.2%), Provinsi Nusa Tenggara Timur (14.8%), Provinsi Maluku (14.3%) dan yang terakhir ditempati Provinsi Jawa Timur (13.5%)
